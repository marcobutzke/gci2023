```python
import pandas as pd
import xlrd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
from sklearn import tree
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from pandas.api.types import is_numeric_dtype
from sklearn.neighbors import NearestNeighbors
```

```python
estados = pd.read_csv('brasil_estados.csv')
estados
```

```python
variaveis = [
    'area',
    'populacao',
    'densidade',
    'matricula',
    'idh',
    'receitas',
    'despesas',
    'rendimento',
    'veiculos'
]
```

```python
data = estados[variaveis].copy()
data['regiao'] = estados['regiao_nome']
X = data.drop(columns='regiao')
Y = data['regiao']
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=3)
decision_tree = decision_tree.fit(X, Y)
text_representation = tree.export_text(decision_tree, feature_names=variaveis)
print(text_representation)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = estados[variaveis].copy()
data['regiao'] = estados['regiao_nome']
X_Train = data.drop(columns=['regiao'], axis=1)
X_Test = data.drop(columns=['regiao'], axis=1)
y_Train = data['regiao']
y_Test = data['regiao']
sc_x = StandardScaler()
X_Train = sc_x.fit_transform(X_Train)
X_Test = sc_x.fit_transform(X_Test)
```

```python
forest = RandomForestClassifier(random_state=0)
forest.fit(X_Train, y_Train)
```

```python
predicao = forest.predict(X_Test)
```

```python
predicao
```

```python
importancia = forest.feature_importances_.tolist()
importancia = pd.DataFrame(importancia)
importancia.columns = ['importances']
importancia['variavel'] = variaveis
importancia
```

```python
data = estados[variaveis].copy()
kmeans = KMeans(n_clusters=5, random_state=0).fit(data)
datac = estados[['regiao_nome','uf']].copy()
datac = datac.merge(data, left_index=True, right_index=True)
datac['cluster'] = kmeans.labels_
datas = datac.sort_values(by='cluster')
datas
```

```python
centroides = kmeans.cluster_centers_
centroides = pd.DataFrame(
    centroides,
    columns = variaveis
)
centroides = centroides.reset_index()
centroides = centroides.rename(columns = {'index': 'cluster'})
centroides
```

```python
data = estados[variaveis].copy()
data['regiao'] = estados['regiao_nome']
X_Train = data.drop(columns=['regiao'], axis=1)
X_Test = data.drop(columns=['regiao'], axis=1)
y_Train = data['regiao']
y_Test = data['regiao']
sc_x = StandardScaler()
X_Train = sc_x.fit_transform(X_Train)
X_Test = sc_x.fit_transform(X_Test)
logreg = LogisticRegression(solver="lbfgs", max_iter=500)
logreg.fit(X_Train, y_Train)
```

```python
pred_logreg = logreg.predict(X_Test)
pred_proba = logreg.predict_proba(X_Test)
```

```python
datar = estados[['regiao_nome','uf']].copy()
datar = datar.merge(data, left_index=True, right_index=True)
datar["regiao_pred"] = pred_logreg
datar
```

```python
lista_proba = pred_proba.tolist()
lista_proba = pd.DataFrame(
    lista_proba,
    columns = [
        'CentroOeste', 'Nordeste', 'Norte', 'Sudeste', 'Sul'
    ]
)
lista_proba
```

```python
datar = pd.merge(datar, lista_proba, left_index=True, right_index=True)
datar
```

```python
data = estados[variaveis].copy()
for estado in estados['uf']:
    print(estado)
    datax = estados[estados['uf'] == estado][variaveis]
    neighbors_alg = NearestNeighbors(n_neighbors=min(5, len(data))).fit(data)
    similar = neighbors_alg.kneighbors(datax, return_distance=False)[0]
    data_similar = estados.iloc[similar][variaveis].reset_index()
    df_uf = estados.iloc[similar]['uf'].reset_index()
    df_uf = df_uf.merge(data_similar, on='index', how='left')
    df_uf = df_uf.drop(columns=['index'])
    print(df_uf)
```
